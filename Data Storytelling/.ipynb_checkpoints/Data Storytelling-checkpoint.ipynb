# Conclusions from investigated trends

In this notebook, I investigated the factors that could be influencing the math, reading and writing scores of students in college. The factors investigated were mainly the gender of the student (M or F), parental level of education, the type of lunch a student has (free/reduced vs. standard), and whether a student had done the test preparation course or not. Given the ambiguity of the race column, that column was not considered in the investigation.

As seen in the initial analysis of the data, the dataset consists of male and female students, with a majority of students being female (36 more females than males out of 1000 students). The majority of students had done no course preparation. Most students had a standard lunch, as opposed to a reduced or free lunch plan. Finally, most students' parental level education was some college or an associate's degree, followed by some high school or completed high school, and a minority had parents with a bachelor or masters degree.

Looking at the distribution of math, reading and writing scores for all students, the grade distributions were approximately normal for all three types of tests (shown in figure 1, figure 2, figure 3, and figure 4), meaning a minority of students had very high or very low grades, and most scored an average grade (between 50-70).

The correlation analysis reveals that students that scored a certain score on any one test were also very likely to achieve a similar score on any of the other tests. This was more true for reading and writing tests (figure 5). Now, we can discuss what it is that causes students to achieve a high score on any of the three tests.

I created additional columns that mark the student as having passed or failed a math, reading or writing test to better help analyze the data. I then created a count plot that counts how many students that have completed the preparation course have passed vs. the number that failed. Since the majority of students did no preparation course work, there are naturally more students that have both passed and failed the three tests that did no course preparation compared ot those that did. However, the ratio of students that failed compared to passed that did complete course prep was lower than those that did not (figure 6, figure 7, and figure 8). This suggests that the course preparation did somewhat help prevent students from failing a math, reading or writing test.

There was no clear relationship between a student's gender and their score on any test. Looking at the count and scatter plots (figure 12, figure 13, figure 14, figure 15, figure 16, and figure 17), females and males had a very similar distribution of scores for all three tests; males or females were not more likely to get lower or higher scores, and both genders scored mostly average scores (between 50-80).

The parents' level of education also did not seem to have a clear effect on a student's test score for math, reading or writing (figure 9, figure 10, figure 11). However, when looking at the relationship between parental level of education, test score, and whether or not a student did the preparation course, the bubble plots show that students with parents who did "some high school" and "high school" had the lowest grades (figure 18, figure 19, and figure 20).

Interestingly, the type of lunch a student was registered for (standard vs. free/reduced) seemed to have some effect on a student's math, reading or writing score. Students with a free or reduced lunch plan were more likely to fail a math, reading or writing test than students with a standard lunch (figure 21, figure 22, and figure 23), and students with a reduced or free lunch had lower average math, reading and writing grades than students with a standard lunch (figure 25, figure 26, and figure 27). 

There was not a clear relationship between type of lunch any given student had and their parents' level of education. The count plot shows that the distribution of education level was very similar for both lunch types. Most parents had "some college" or an associates degree, and the least common was a bachelor or masters degree, for both lunch types (figure 24). 


To conclude, after investigating all factors in this dataset that affect student grades (excluding race/ethnicity), it appears that the strongest influencer of grades was the type of lunch a student had, which is an economic factor, followed by whether a test preparation course was done, a more personal factor (it could also be an economic factor, but that cannot be assumed because the price of the test preparation course is not known). The gender of a student and the parental educational level did not seem to have any real effect on student test scores.


Given these conclusions, a good question to further investigate, and possibly can be used to improve students' grades in the future, is what other economic factors could be involved in affecting student grades? We can start by looking into the price of the test preparation course, and how to better assist students and parents with funding it.
