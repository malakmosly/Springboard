```python
df['math pass status'] = np.where(df['math score'] < 50, 'F', 'P')
df['reading pass status'] = np.where(df['reading score'] < 50, 'F', 'P')
df['writing pass status'] = np.where(df['writing score'] < 50, 'F', 'P')
```

```python
df.head()
```

```python
sns.countplot(data = df, x = 'test preparation course', hue = 'math pass status')
plt.show()
```

```python
sns.countplot(data = df, x = 'test preparation course', hue = 'reading pass status')
plt.show()
```

```python
sns.countplot(data = df, x = 'test preparation course', hue = 'writing pass status')
plt.show()
```

```python
sns.countplot(data = df, x = 'parental level of education', hue = 'math pass status')
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.countplot(data = df, x = 'parental level of education', hue = 'reading pass status')
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.countplot(data = df, x = 'parental level of education', hue = 'writing pass status')
plt.xticks(rotation = 45)
plt.show()
```
